## Loading the LLM


```python
import transformers
import torch

model_id = "HPAI-BSC/Llama3-Aloe-8B-Alpha"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto"
)
```

## These are the different html blocks that can be used as the input to test the model.

```python
html_block1 = '''
<div class="product">
    <div class="product-image">
        <img src="https://images-cdn.ubuy.co.in/63402ef469382d3105512dcf-royal-brown-basmati-rice-32-oz.jpg" alt="Product Image">
    </div>
    <div class="product-details">
        <h2 class="product-name">907gm Brown Basmati Rice</h2>
        <p class="product-price">$24.56</p>
        <p class="product-description">Long grains, Nice Aroma, Cooks Fast</p>
    </div>
</div>
'''

html_block2='''
<div class="product-card">
    <div class="product-image">
        <img src="https://example.com/images/smartphone.jpg" alt="Smartphone">
    </div>
    <div class="product-info">
        <h2 class="product-name">Latest Smartphone</h2>
        <p class="product-price">$799.99</p>
        <p class="product-description">High-resolution display, fast processor, and long battery life.</p>
    </div>
    <div class="product-actions">
        <button class="add-to-cart">Add to Cart</button>
        <button class="buy-now">Buy Now</button>
    </div>
</div>
'''

html_block3='''
<div class="product-listing">
    <div class="item">
        <img class="item-image"
        src="https://bhotahiti-bucket.s3.ap-south-1.amazonaws.com/products/products/166651834531166.webp"
        alt="Product 1 Image">
        <div class="item-info">
            <h3 class="item-name">500g Organic Quinoa</h3>
            <span class="item-price">$15.99</span>
            <p class="item-description">High in protein, Gluten-free, Non-GMO</p>
        </div>
    </div>
    <div class="item">
        <img class="item-image" src="https://m.media-amazon.com/images/I/919OdYC6uKL.jpg" alt="Product 2 Image">
        <div class="item-info">
            <h3 class="item-name">1kg Whole Wheat Flour</h3>
            <span class="item-price">$8.50</span>
            <p class="item-description">Stone-ground, Organic, Rich in fiber</p>
        </div>
    </div>
</div>
'''
```

## Making sure the model understands what we are trying to acheive

```python
questions=['What is the name of the product?',
           'What is the price of that product?',
           'A brief description about the product?',
           'is there any image associated with that product?']

answer_direction=f"Please provide answer of the questions in a json format by analyzing the html content provided. also provide their corresponding CSS selectors.\
                  Look for multiple answers to the same questions. for example: there can be multiple products available \
                  in the html block. if such case arise, add it to the list as given in the example."


example='''{
    "product-name":[{
        "selector": replace with CSS selector,
        "answer": replace with actual name of the product as provided in html content
    }],
    "product-price":[{
        "selector": replace with CSS selector,
        "answer": replace with price of the product as provided in html content
    }],
    "product-description":[{
        "selector": replace with CSS selector,
        "answer": replace with product description as provided in html content
    }],
    "product-image":[{
        "selector": replace with CSS selector,
        "answer": replace with image link if available in html content
    }]
}'''

messages = [
    {"role": "system", "content": f"You are a expert system at analyzing the hypertext markup langugage and answering question on such context.\
                                    Here is a small html content{html_block1} for you to analyze."},
    {"role": "user", "content": f"Here are multiple questions you need to answer {questions}. {answer_direction}\
                                  here is an example for you. {example}.Do not generate anything else ,not even a character other than the json output \
                                  starting with a curly brace so that i can access the answer easily."},
]
```

## Generating Output ( will take an average time of 2 minutes to execute)


```python
prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9
)
```
